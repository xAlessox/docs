```python
##---------------------PARTE 1
"""Generar un obejto Series que presente las ventas de los últimos 5 días"""
```

```python
##Solicitando al usuario las ventas de los últimos 5 días y almacenandolo en una lista

print("Ingrese las ventas de los últimos 5 días...\n")

venta_diaria = []

for dia in range(5):

  while True:

    try:
      print("Ingrese el monto total vendido en el día",dia+1,":")
      venta =(float(input()))

      if type(venta) == float and venta > 0:
        venta_diaria.append(venta)
        break
      else:
        print("Ingrese un valor válido")
    except:
      print("Ingrese un valor válido")
```

```python
##Importando el paquete pandas

import pandas as pd

##Creando un objeto "Series" con los datos de las ventas de los últimos 5 días

series_ventadiaria = pd.Series(venta_diaria)
series_ventadiaria
```

```python
##---------------------PARTE 1
"""Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:"""
```

```python
##Creando listas con los datos de la tabla mostrada en la guia

lmeses = ['Enero','Febrero','Marzo','Abril','Mayo']
lventas = [30500,35600,28300,33900,42500]
lgastos = [22000,23450,18100,35700,32450]

ldatos = [lventas,lgastos]

##Importanto el paquete pandas para generar dataframe

import pandas as pd

##Creando DataFrame
index = ['Ventas','Gastos']

df_datos = pd.DataFrame(ldatos, index = index, columns = lmeses).transpose()
df_datos

##Agregando las nuevas columnas

```

```python
##---------------------PARTE 2
"""A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril"""
```

```python
##Importando paquete NumPy para crear las nuevas columnas a partir de operaciones matemáticas

import numpy as np

##Creando arrays a partir de las listas

np_lmeses = np.array(lmeses)
np_lventas = np.array(lventas)
np_lgastos = np.array(lgastos)

##Creando nuevas columnas a partir de los calculos

columna1 = (np_lgastos/np_lventas) * 100
columna2 = np.where(np_lventas >= 30000,"META SUPERADA","META NO SUPERADA")
columna3 = np.where(np_lgastos > np_lventas, np_lgastos - np_lventas, 0)

##Incorporando nuevas columnas

df_datos["columna1"] = columna1
df_datos["columna2"] = columna2
df_datos["columna3"] = columna3

df_datos
```

```python
##Obteniendo el total de las ventas en los ultimos 3 meses

df_ventaslast3m = df_datos.tail(3)
ventaslast3m = np.sum(df_ventaslast3m["Ventas"])
print("El total de ventas en los ult 3 meses es:",ventaslast3m)

##Obteniendo el total de gastos de febrero a abril

fila_gastos = ["Febrero","Marzo","Abril"]

df_gastosmeses = np.sum(df_datos.loc[fila_gastos,"Ventas"])
df_gastosmeses
print("El total de gastos de febrero a abril es:",df_gastosmeses)
```
